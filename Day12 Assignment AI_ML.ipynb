# Problem Statement 1
IF P VALUE < 0.05, REJECT Ho
IF P VALUE > 0.05, ACCEPT HO

1st Sample- Employee with Attrition as "Yes" with certain age values.

2nd Sample- Employee with Attrition as "No" with certain age values.

Both these samples are independent so we have to apply Mann Whitney Test

```python
from scipy.stats import mannwhitneyu
stats1,p1=mannwhitneyu(dataset3.Age,dataset4.Age)
print(stats1,p1)
```

H0: There is no significant differnce in the employee's age between Attrition(Yes) and Attrition(No).(Null hypothesis)

H1: There is significant differnce in the employee's age between Attrition(Yes) and Attrition(No).(Alternate hypothesis)

# Problem Statement 2
1st Sample- Employee with Attrition as "Yes" with certain distance from home values.

2nd Sample- Employee with Attrition as "No" with certain distance from home values.

Both these samples are independent so we have to apply Mann Whitney Test

```python
from scipy.stats import mannwhitneyu
stats2,p2=mannwhitneyu(dataset3.DistanceFromHome,dataset4.DistanceFromHome)
print(stats2,p2)
```

H0: There is no significant differnce in distance from home between Attrition(Yes) and Attrition(No).(Null hypothesis)

H1: There is significant differnce in distance from home between Attrition(Yes) and Attrition(No).(Alternate hypothesis)

The p value>0.05 so null hypothesis is accepted and alternate hypothesis is rejected
